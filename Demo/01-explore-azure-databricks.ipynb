### 🎯 DEMO: Creating an Azure Databricks Workspace

You can create an Azure Databricks workspace using multiple methods. Let's explore the Azure CLI and PowerShell approaches.

**Using Azure CLI:**

```bash
# Create a resource group (if needed)
az group create --name myresourcegroup --location westus2

# Create an Azure Databricks workspace
az databricks workspace create \
    --resource-group myresourcegroup \
    --name mydatabricksws \
    --location westus2 \
    --sku premium
```

**Using PowerShell:**

```powershell
# Create an Azure Databricks workspace
New-AzDatabricksWorkspace `
    -Name mydatabricksws `
    -ResourceGroupName myresourcegroup `
    -Location westus2 `
    -ManagedResourceGroupName databricks-managed-rg `
    -Sku premium
```

**Key Parameters:**
- **Pricing Tier**: Standard (basic features), Premium (RBAC, Unity Catalog), or Trial (14-day free)
- **Managed Resource Group**: Auto-created group for Databricks infrastructure resources

### 🎯 DEMO: SQL Workload

Databricks SQL enables data analysts to query data using familiar SQL syntax, create dashboards, and perform BI analytics.

**Note:** SQL warehouses are available in Premium tier and above.

````python
%sql
-- SQL Analytics Example: Query Delta tables

-- Create a sample analysis query
SELECT 
    customer_region,
    YEAR(order_date) as order_year,
    COUNT(*) as total_orders,
    SUM(order_amount) as total_revenue,
    AVG(order_amount) as avg_order_value
FROM trainer_demo.demo_01.orders
WHERE order_date >= '2024-01-01'
GROUP BY customer_region, YEAR(order_date)
ORDER BY total_revenue DESC;

-- This query can be:
-- 1. Run in the SQL Editor and/or in this notebook
-- 2. Saved as a query for reuse
-- 3. Added to dashboards
-- 4. Scheduled to run automatically
````
